### Task 01: Exploring and Visualizing Iris Dataset

```python
# imorting libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = sns.load_dataset("iris")
```

### Exploration

```python
print("First 5 rows:\n", df.head())
print("Last 5 rows:\n", df.tail())
```

```python
print("Shape:", df.shape)
```

```python
# Columns
print("Columns:", df.columns.tolist())
```

```python
# Summary stataistics
print("Description:\n", df.describe())
print("Information of Data:\n", df.info())
```

```python
# checking for the null values
print("Null value is: ", df.isnull().sum())
```

### Visualization

##### Scatter Plot

```python
# Scatter plot (sepal_length vs sepal_width, colored by species)
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species',  palette='viridis')
plt.title("Sepal Length vs Width by Species")
plt.xlabel("sepal_length")
plt.ylabel("Sepal_width")
plt.show()
```

##### Histogram

```python
# Histograms
df.hist(figsize=(10, 6), bins=10, color="skyblue", edgecolor="black")
plt.suptitle("Histograms of Iris Features")
plt.show()
```

##### Box Plot

```python
# Box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette='coolwarm')
plt.title("Box Plot of Iris Features")
plt.show()
```

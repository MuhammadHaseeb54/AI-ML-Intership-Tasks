### Task: Customer Churn Pipliner Generator 

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report
from joblib import dump
from pipeline import load_data, build_pipeline
```

```python
# Load data
df = load_data("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
X = df.drop("Churn", axis=1)
y = df["Churn"].map({'Yes': 1, 'No': 0})  # Convert labels to binary
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Build pipeline with base model
pipeline = build_pipeline(model='rf')

# Hyperparameter grid
param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5],
}
```

```python
# GridSearch
grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Save model
dump(grid_search.best_estimator_, "churn_pipeline.pkl")
```

```python
# Evaluate
y_pred = grid_search.predict(X_test)
print(classification_report(y_test, y_pred))
```
